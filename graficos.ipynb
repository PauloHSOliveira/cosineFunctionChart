```
import math
import matplotlib.pyplot as plt

x=[]
for i in range(-80, 81, 1):
  i=i/10
  x.append(i)
print(x)

cosseno = [0]*len(x)
for i in range(len(x)):
  cosseno[i]=math.cos(2*math.pi*x[i])

print(cosseno)

plt.figure(1)
plt.plot(x, cosseno)
plt.title("Função cosseno x")
plt.xlabel("x")
plt.ylabel("y(x)")

plt.show()
```

```
from random import randint
import matplotlib.pyplot as plt
import numpy as np

mu, sigma = 0, 0.1
s = np.random.normal(mu, sigma, 1000)

#Verificando a média e a variância

abs(mu - np.mean(s)) < 0.01
abs(sigma - np.std(s, ddof=1)) < 0.01

# Construindo o histograma juntamente com a distribuição de probabilidade
plt.figure(1)

# Construindo o histograma juntamente com a distribuição de probabilidade
count, bins, ignored = plt.hist(s, 30, normed=True)
plt.plot()
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 /
(2 * sigma**2) ), linewidth=2, color='r')

plt.show()
```
