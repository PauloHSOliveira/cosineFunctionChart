```
import matplotlib.pyplot as plt
import math
import numpy

x = []
for i in range(-80, 81, 1):
  i=i/10
  x.append(i)
print(x)

cosseno = [0]*len(x)

for i in range(len(x)):
  cosseno[i]=math.cos(2*math.pi*x[i])
print(cosseno)

plt.figure(1)
plt.plot(x, cosseno)
plt.title('Função cosseno X')
plt.xlabel('x')
plt.ylabel('y(x)')

plt.show()
```
